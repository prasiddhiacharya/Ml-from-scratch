```python
#examine the datatypes
data_types = df.dtypes
data_types
```

```python
# identifying null values variable
isnull = df.isnull()
isnull
```

```python
isnull_count = isnull.sum()
print("Total number of null values:", isnull_count)
```

```python
# total nulls in dataset
total_isnull_count = isnull.sum() .sum()
print("Total nulls:", total_isnull_count)
```

```python
# replace single quote with blank and make the date formatted
df["Date"] = df["Date"].replace("'"," ", regex = True)
df
```

```python
# handle null values, return dataframe with no empty cells
# we can use df.dropna(), it removes empty cells but will not change the dataframe
#if we want data frame to be changed, simply use inplace= True
new_df = df.dropna()
new_df
```

```python
total_isnull_count_newdf = new_df.isnull().sum() .sum()
print("Total nulls:", total_isnull_count)
```

```python
# another way of handling the null values is by replacing the null values
df.fillna(130, inplace =True)
df
```

```python
df = pd.read_csv("/content/sample_data/pandas_dataset.csv")
df["Calories"].fillna(130, inplace=True)
df
```

```python
# fill specific with mean, median and mode
df = pd.read_csv("/content/sample_data/pandas_dataset.csv")
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace= True)
df
```
